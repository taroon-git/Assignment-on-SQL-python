Q1: What is databases ? what is the difference between SQL and NoSQL ?

```python
Databaase:- 

A database is an organized collection of structured information, or data,
typically stored electronically in a computer system.
A database is usually controlled by a database management system (DBMS).

SQL :- 

1. Relational database management system
2. It has static schema
3. It is vertically scalable
4. It follows ACID property
5. Ex- Ms-SQL , Oracle etc.

NoSQL :-  

1. Non- relational or distributed database system
2. They have dynamic schema
3. It is Horizontally scalable 
4. It doesn't follows ACID property(It follows CAP)
5. ex- MongoDB etc.
```

Q:2  What is DDL ? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

```python
DDL:- (Data Definition Language)

DDL refers to a language which is used to modify data and define data structures.
The DDL commands used to :-  remove, add, or modify tables within a database.

CREATE :- 

It has a predefined syntax.
CREATE is used to build a  new table.

syntax:- 
CREATE TABLE [table_name] ([definition]);

ALTER:- 

This command modifies any existing table in a database.
This command can add columns , drop the existing columns and 
even change the datatypes of columns .

syntax:- 
ALTER TABLE type_of_object name_of_object parameters;


DROP:- 

The DROP command is used in order to delete objects, like table
index etc. 
If once a certain object is destroyed then, can't recover.

syntax:-  
DROP type_of_object name_of_object;

TRUNCATE:- 

It is also just like a DROP statement, It is mainly used to remove 
all the records from the table quickly.
It is just like a DROP statement but TRUNCATE statement preserve
teh full structure so as to be reused later.

syntax:- 
TRUNCATE TABLE name_of_table
```

Q:5 Explain Primary key and Foreign key?

 Primary Key:-

               A primary key is used to ensure that data in the specific column is unique. A column cannot have NULL values.
               It is either an existing table column or a column
               that is specifically generated by the database according to a defined sequence. 

 Foreign key:- 

    A Foreing key is a column or group of columns in a relational database table that provides a link between data in two tables.
             It is a column (or columns) that references a column (most often the primary key) of another table.
